```python
import pygame
import numpy as np
import random
import cv2
import os
os.mkdir('frame')
del os
```

```python
class ai:
    def __init__(self,n_row=3,n_column=4,goal_r=0,goal_c=3,n_fire=1,n_wall=1,start_r=2,start_c=0,goal_reward=1,fire_reward=-1,cell_reward=0,cell_side=100,n_games=10,move_decision_factor=1,increment_factor=2,max_move_decision_factor=40,max_step=200):
        self.n_row=n_row
        self.n_column=n_column
        self.goal_r=goal_r
        self.goal_c=goal_c
        self.n_fire=n_fire
        self.n_wall=n_wall
        self.start_r=start_r
        self.start_c=start_c
        self.goal_reward=goal_reward
        self.fire_reward=fire_reward
        self.cell_reward=cell_reward
        self.cell_side=cell_side
        self.n_games=n_games
        self.move_decision_factor=move_decision_factor
        self.increment_factor=increment_factor
        self.max_move_decision_factor=max_move_decision_factor
        self.steps=0
        self.current_x=start_r
        self.current_y=start_c
        self.games_played=0
        self.max_step=max_step
        self.wall_coord=np.empty((n_wall,2))
        self.next_x=start_r
        self.next_y=start_c
        self.last_move=-1
        
        for i in range(n_wall):
            self.wall_coord[i]=np.array(list(map(int,input().split())),dtype=int)

        self.fire_coord=np.empty((n_fire,2))
        for i in range(0,n_fire):
            self.fire_coord[i]=np.array(list(map(int,input().split())),dtype=int)
                #assigning reward to each state
        self.grid=np.zeros((n_row,n_column,6),dtype=float)
        for i in range(n_row):
            for j in range(n_column):
                self.grid[i][j][4]=cell_reward
        self.grid[goal_r][goal_c][4]=goal_reward
        self.grid[goal_r][goal_c][5]=goal_reward

        for i in range(n_fire):    
            self.grid[int(self.fire_coord[i][0])][int(self.fire_coord[i][1])][4]=fire_reward    
            self.grid[int(self.fire_coord[i][0])][int(self.fire_coord[i][1])][5]=fire_reward    
    
    def update(self):
        lt=[0]*int(self.move_decision_factor)+[1]*int(self.max_move_decision_factor-self.move_decision_factor)
        move_type=random.choice(lt)
        del lt
      
        if move_type==1:
            move=random.randint(0,3)
        else:
            if self.grid[self.current_x][self.current_y][0]>=self.grid[self.current_x][self.current_y][1]and self.grid[self.current_x][self.current_y][0]>=self.grid[self.current_x][self.current_y][2] and self.grid[self.current_x][self.current_y][0]>=self.grid[self.current_x][self.current_y][3]:
                move=0
            elif self.grid[self.current_x][self.current_y][1]>=self.grid[self.current_x][self.current_y][0]and self.grid[self.current_x][self.current_y][1]>=self.grid[self.current_x][self.current_y][2] and self.grid[self.current_x][self.current_y][1]>=self.grid[self.current_x][self.current_y][3]:
                move=1
            elif self.grid[self.current_x][self.current_y][2]>=self.grid[self.current_x][self.current_y][0]and self.grid[self.current_x][self.current_y][2]>=self.grid[self.current_x][self.current_y][1] and self.grid[self.current_x][self.current_y][2]>=self.grid[self.current_x][self.current_y][3]:
                move=2
            else:
                move=3
        
        self.next_x=self.current_x
        self.next_y=self.current_y
        if move==0 and self.last_move!=2:
            self.next_x-=1
            if self.next_x<0 or (np.array([self.next_x,self.next_y]) == self.wall_coord).all(1).any() :
                self.next_x+=1
        elif move==1 and self.last_move!=3:
            self.next_y+=1
            if self.next_y>=self.n_column or (np.array([self.next_x,self.next_y]) == self.wall_coord).all(1).any():
                self.next_y-=1
        elif move==2 and self.last_move!=0:
            self.next_x+=1
            if self.next_x>=self.n_row or (np.array([self.next_x,self.next_y]) == self.wall_coord).all(1).any():
                self.next_x-=1
        elif move==3 and self.last_move!=1:
            self.next_y-=1
            if self.next_y<0 or (np.array([self.next_x,self.next_y]) == self.wall_coord).all(1).any():
                self.next_y+=1
        
        self.last_move=move
        

        if [self.next_x,self.next_y]!=[self.current_x,self.current_y]:
        
            self.grid[self.current_x][self.current_y][move]=self.grid[self.next_x][self.next_y][4]+0.9*self.grid[self.next_x][self.next_y][5]-0.1*self.grid[self.current_x][self.current_y][move]
            self.grid[self.current_x][self.current_y][5]=max([self.grid[self.current_x][self.current_y][i] for i in range(4)])

        self.current_x=self.next_x
        self.current_y=self.next_y
        if (np.array([self.next_x,self.next_y]) == self.fire_coord).all(1).any() or [self.goal_r,self.goal_c]==[self.next_x,self.next_y] or self.steps>=self.max_step:
            self.games_played+=1
            self.move_decision_factor+=self.increment_factor
            self.current_x=self.start_r
            self.current_y=self.start_c
            self.steps=0
            self.last_move=-1
            
        
        
        self.steps+=1
        if self.move_decision_factor>=self.max_move_decision_factor:
            self.move_decision_factor=self.max_move_decision_factor
    
    def save_frame(self,screen,i):
        
        pygame.image.save(screen,'frame/'+str(i)+'.jpg')
                

    
    def game_screen(self,screen):

        #to display ai position
        pygame.draw.rect(screen,(120,100,0),(self.n_column*self.cell_side+100+self.current_y*self.cell_side,self.current_x*self.cell_side,self.cell_side,self.cell_side))
        #to diplay direction
        font = pygame.font.Font('freesansbold.ttf', 20)
        for i in range(self.n_row):
            for j in range(self.n_column):
                
                if self.grid[i][j][0]==self.grid[i][j][5]:

                    text = font.render('UP', True, (0,0,0), (255,255,255))
                    textRect = text.get_rect() 
                    textRect.center = ((self.n_column+j+1.5)*self.cell_side,(i+0.5)*self.cell_side)
                    screen.blit(text, textRect) 
                elif self.grid[i][j][1]==self.grid[i][j][5]:

                    text = font.render('RIGHT', True, (0,0,0), (255,255,255))
                    textRect = text.get_rect() 
                    textRect.center = ((self.n_column+j+1.5)*self.cell_side,(i+0.5)*self.cell_side)
                    screen.blit(text, textRect) 
                elif self.grid[i][j][2]==self.grid[i][j][5]:

                    text = font.render('DOWN', True, (0,0,0), (255,255,255))
                    textRect = text.get_rect() 
                    textRect.center =((self.n_column+j+1.5)*self.cell_side,(i+0.5)*self.cell_side)
                    screen.blit(text, textRect) 
                else:

                    text = font.render('LEFT', True, (0,0,0), (255,255,255))
                    textRect = text.get_rect() 
                    textRect.center = ((self.n_column+j+1.5)*self.cell_side,(i+0.5)*self.cell_side)
                    screen.blit(text, textRect) 
        #to diplay fire
        for i in range(self.n_fire):
            pygame.draw.rect(screen,(255,69,0),(self.n_column*self.cell_side+100+self.fire_coord[i][1]*self.cell_side,self.fire_coord[i][0]*self.cell_side,self.cell_side,self.cell_side))
        #to display wall
        for i in range(self.n_wall):
            pygame.draw.rect(screen,(100,100,100),(self.n_column*self.cell_side+100+self.wall_coord[i][1]*self.cell_side,self.wall_coord[i][0]*self.cell_side,self.cell_side,self.cell_side))
        #to display goal
        pygame.draw.rect(screen,(0,255,0),(self.n_column*self.cell_side+100+self.goal_c*self.cell_side,self.goal_r*self.cell_side,self.cell_side,self.cell_side))
        
        for i in range(self.n_row+1):
            pygame.draw.line(screen,(0,0,0),(self.n_column*self.cell_side+100+0,i*self.cell_side),(self.n_column*self.cell_side+100+self.n_column*self.cell_side,i*self.cell_side),2)
        for i in range(self.n_column+1):
            pygame.draw.line(screen,(0,0,0),(self.n_column*self.cell_side+100+i*self.cell_side,0),(self.n_column*self.cell_side+100+i*self.cell_side,self.n_row*self.cell_side),2)

    def q_screen(self,screen):

        for i in range(self.n_row+1):
            pygame.draw.line(screen,(0,0,0),(0,i*self.cell_side),(self.n_column*self.cell_side,i*self.cell_side),2)
        for i in range(self.n_column+1):
            pygame.draw.line(screen,(0,0,0),(i*self.cell_side,0),(i*self.cell_side,self.n_row*self.cell_side),2)
        font = pygame.font.Font('freesansbold.ttf', 30)

        text = font.render(str(self.games_played+1), True, (0,0,0), (255,255,255))
        textRect = text.get_rect() 
        textRect.center = ((self.n_column+0.5)*self.cell_side,(self.n_row//2)*self.cell_side)
        screen.blit(text, textRect) 
        font = pygame.font.Font('freesansbold.ttf', 15)
        for i in range(self.n_row):
            for j in range(self.n_column):
                pygame.draw.line(screen,(0,0,0),(j*self.cell_side,i*self.cell_side),((j+1)*self.cell_side,(i+1)*self.cell_side),1)
                pygame.draw.line(screen,(0,0,0),(j*self.cell_side,(i+1)*self.cell_side),((j+1)*(self.cell_side+1),i*(self.cell_side+1)),1)

                text = font.render('{:.4f}'.format(self.grid[i][j][0]), True, (0,0,0), (255,255,255))
                textRect = text.get_rect() 
                textRect.center = ((j+0.5)*self.cell_side,(i+0.25)*self.cell_side)
                screen.blit(text, textRect) 

                text = font.render('{:.4f}'.format(self.grid[i][j][1]), True, (0,0,0), (255,255,255))
                textRect = text.get_rect() 
                textRect.center = ((j+0.75)*self.cell_side,(i+0.5)*self.cell_side)
                screen.blit(text, textRect) 

                text = font.render('{:.4f}'.format(self.grid[i][j][2]), True, (0,0,0), (255,255,255))
                textRect = text.get_rect() 
                textRect.center = ((j+0.5)*self.cell_side,(i+0.75)*self.cell_side)
                screen.blit(text, textRect) 

                text = font.render('{:.4f}'.format(self.grid[i][j][3]), True, (0,0,0), (255,255,255))
                textRect = text.get_rect() 
                textRect.center = ((j+0.25)*self.cell_side,(i+0.5)*self.cell_side)
                screen.blit(text, textRect) 



```

```python
game=ai(n_row=3,n_column=4,goal_r=0,goal_c=3,n_fire=1,n_wall=1,start_r=2,start_c=0,goal_reward=2,fire_reward=-1,cell_reward=0,cell_side=100,n_games=20,move_decision_factor=1,increment_factor=0,max_move_decision_factor=40,max_step=200)
```

```python
pygame.init()
screen=pygame.display.set_mode((game.n_column*game.cell_side*2+100,game.n_row*game.cell_side))
pygame.display.set_caption('finding goal-q_value')
run=True
start=False
while run:
    if start:
        screen.fill([255,255,255])

        # for q values
        game.q_screen(screen)

        #for game view
        game.game_screen(screen)
        if game.games_played<=game.n_games:
            game.update()

        else:
            run=False



        for event in pygame.event.get():

            if event.type==pygame.QUIT:
                run=False
        pygame.display.update() 
        pygame.time.delay(100)
    else:
        for event in pygame.event.get():

            if event.type==pygame.KEYDOWN:
                start=True

        
pygame.quit()
```
